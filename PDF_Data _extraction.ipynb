# Text, Image extraction of typeset/actual PDF



```python
#Extract text from PDF
import PyPDF2 
 
# creating a pdf file object 
pdfFileObj = open("Data/11230_2004_Article_5114431.pdf", 'rb') 
 

pdfReader = PyPDF2.PdfReader(pdfFileObj)

pageObj = pdfReader.pages
 
for page in pageObj:
    print(page.extract_text()) 
 
```

```python
# Extract images from PDF
import PIL.Image
from PIL import Image
import fitz
import io
import matplotlib.pyplot as plt

filename = "11230_2004_Article_5114431"
pdf = fitz.open("Data/11230_2004_Article_5114431.pdf") 
counter = 1
for i in range(len(pdf)):
    page = pdf[i]
    images= page.get_images()
    for image in images:
        base_img = pdf.extract_image(image[0])
        image_data = base_img["image"]
        img = PIL.Image.open(io.BytesIO(image_data))
        extension = base_img["ext"]
        img.save(open(f"{filename}-Image{counter}.{extension}", "wb"))
        counter +=1
```
