```python
import xml.etree.ElementTree as et
import os
```

```python
book_xml_folder = "output/books"
if not os.path.exists(book_xml_folder):
    print("Error: Manifest folder does not exists.")
```

```python
#Extracting only 1000 book records for Data analysis
books_data = [os.path.join(book_xml_folder, file) for file in os.listdir(book_xml_folder)]
books_data
```

```python
# Out of total book record how many have chapter data
chapter_substring = "Chapter"
BookFrontMatter_substring = "BookFrontMatter"
BookBackMatter_substring = "BookBackMatter"
book_substring = "Book_nlm"

```

```python
chapter = 0
BookFrontMatter = 0
BookBackMatter = 0
book = 0

def count_book_details(xml):
    global chapter, BookFrontMatter, BookBackMatter, book
    if chapter_substring in xml:
        chapter += 1        
    
    if BookFrontMatter_substring in xml:    
        BookFrontMatter += 1        

    if BookBackMatter_substring in xml:
        BookBackMatter += 1        

    if book_substring in xml:
        book += 1       
```

```python
book_count = len(books_data)
abstract_count = 0
body_count = 0



for xml in books_data:
    count_book_details(xml)
    tree = et.parse(xml)
    root = tree.getroot()
    if root.find("./book-part/body/book-part/book-part-meta/abstract"):
        abstract_count += 1
    if root.find("./book-part/body/book-part/body"):
        body_count += 1
```

```python
print(f"Out of total {book_count} book records, {chapter} are chapters, {BookFrontMatter} are BookFrontMatter,{BookBackMatter} are BookBackMatter and {book} are books end pages")
```
